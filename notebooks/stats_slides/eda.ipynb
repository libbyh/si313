```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load a dataset
df = pd.read_csv('../../data/fortnite_players_stats.csv')

# See the first few rows
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize a distribution
sns.histplot(df['Solo minutesPlayed'], kde=True)
plt.show()
```

```python
# Plot the relationship between two variables with a scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Solo minutesPlayed', y='Solo kills')
plt.title('Kills vs. Minutes Played (Solo mode)')
plt.show()
```
